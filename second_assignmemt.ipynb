```python
# solution of seventh question
data = (5, 10, 15, 20, 25, 30, 35)
count = 0
sum = 0
for i in data:
    if i % 5 == 0:
        count += 1
    sum += i
print('count of element divisble by 5 is :',count)
print('sum of all elements is :',sum)
print('average of tuple is :',sum/len(data))
```

```python
# solution of eighth question
students = {
"Aman": 78,
"Riya": 92,
"Kriti": 88,
"Rahul": 95
}
high = 0
low = 100
count = 0
for i in students.values():
    if i < low:
        low = 1
    if i > high:
        high = i
    if i > 85:
        count += 1
print('highest marks is :',high)
print("lowest marks is :",low)
print('count of the students who got more then 85 is :',count)
```

```python
# solution of ninth question
def freq(lst):
    my_dict = {}
    for i in lst:
        if i not in my_dict:
            my_dict[i] = lst.count(i)
    return my_dict
lst = [1,2,3,43,2,1,3,4,2]
frq = freq(lst)
for i,j in frq.items():
    print(i,'->',j)
```
