```python
# solution of fourth question
# upper() it changes the string to upper case
text = "python"
print(text.upper())

# lower() it changes the string to lower case
text = "PYTHON"
print(text.lower())

# strip() it removes the whitespaces from start and end
text = "  Hello  "
print(text.strip())

# replace() it replaces a substring of a string from another substring
text = "Hello World"
print(text.replace("World", "Python"))

# split() it splits the string into substring on a particular character
fruits = "apple,banana,mango"
print(fruits.split(","))
```
